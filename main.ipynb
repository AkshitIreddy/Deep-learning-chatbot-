```python
import json 
import numpy

#load the dataset
with open('intents.json') as f:
    data = json.load(f)

#This contains all the training sentences 
dataset = []

#This is the list of labels for each sentence in the dataset(same label may repeat)
dataset_labels = []

#These are the different types of labels in the dataset(unique labels only)
tags = []

#List of responses that the chatbot might give for a specific label
responses = []

for intent in data["intents"] :
    for pattern in intent["patterns"] :
        dataset.append(pattern)
        dataset_labels.append(intent['tag'])
    responses.append(intent['responses'])
    
for i in dataset_labels :
    if i not in tags:
        tags.append(i)
        
print("The dataset now looks like this")        
print(dataset)

print("\nThe dataset_labels now looks like this")    
print(dataset_labels)

print("\nThe tags now looks like this")    
print(tags)

print("\nThe responses now looks like this")    
print(responses)
```

```python
#Count the number of tags
tag_count = len(tags)
```

```python
#Convert the dataset_labels from words to numbers
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
dataset_labels = label_encoder.fit_transform(dataset_labels)
```

```python
#Tokenize the words in training sentences and add padding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words= 500 , oov_token= "<OUT>")
tokenizer.fit_on_texts(dataset)

tokenized_sentences = tokenizer.texts_to_sequences(dataset)
padded_sentences = pad_sequences(tokenized_sentences , truncating='post', maxlen= 20)
```

```python
#Create the model
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D

model = Sequential()
model.add(Embedding(1000, 15, input_length = 20))
model.add(GlobalAveragePooling1D())
model.add(Dense(32 , activation = "relu"))
model.add(Dense(32 , activation = "relu"))
model.add(Dense(tag_count , activation = "softmax"))

model.compile(loss = "sparse_categorical_crossentropy" , optimizer = "adam" , metrics = ["accuracy"])
model.summary()
```

```python
history = model.fit(padded_sentences, numpy.array(dataset_labels), epochs= 500)
```

```python
#Save the trained model
model.save("chatbot_model")

import pickle
#Save the fitted tokenizer
with open('tokenizer.pickle', 'wb') as f:
    pickle.dump(tokenizer, f)
    
#Save the fitted label encoder
with open('label_encoder.pickle', 'wb') as f:
    pickle.dump(label_encoder, f)
```
